```python
# Creating a new column to calculate the percentage weighting of each stock in the S&P500

Rawdata = Rawdata.replace(to_replace='NA', value = 0)
Rawdata['Market Cap'] = Rawdata['Market Cap'].astype(float)
Rawdata['S&P500 Market Cap'] = Rawdata['S&P500 Market Cap'].astype(float)
```

```python
print(Rawdata['Market Cap'])
```

```python
print(Rawdata['S&P500 Market Cap'])
```

```python
Rawdata['S&P500 Weighting %'] = Rawdata['Market Cap'] / Rawdata['S&P500 Market Cap'] * 100
print(Rawdata['S&P500 Weighting %'])
```

```python
# New column added which is the calculated weighting of each stock listed on the S&P500.

list(Rawdata)
```

```python
# The next step is to group the companies by sector to determine the sector which is the most heavily weighted in the S&P500

Rawdata.groupby(['Sector'])['S&P500 Weighting %'].sum()
```

```python
# The below is a graph representation of the above groupby function.

Grouped_Data = Rawdata.groupby(['Sector'])['S&P500 Weighting %'].sum().reset_index()
print(Grouped_Data)
Rawdata['S&P500 Weighting %'] = Rawdata['S&P500 Weighting %'].astype(float)
```

```python
Bar = sb.barplot(x="S&P500 Weighting %", y = "Sector", data = Grouped_Data)
sb.set(rc = {'figure.figsize':(30,14)})
Bar.set_xlabel("S&P500 Weighting %")
Bar.set_ylabel("Sector")
Bar.set_title("S&P500 Weightings per sector")
```

```python
# Importing the Yahoo! Finance API and sourcing S&P500 Data
import yfinance as yf
SP_500 = yf.download("^GSPC", start="1990-01-01", end="2021-05-13")
SP500_Data = SP_500.sort_values(by=['Date'], ascending=False)
SP500_Data.head(10)
```

```python
# The API data has sourced the relevant data of open, high, low & closing prices.

SP_500['Close'].plot(title='S&P 500 Closing Price 1990-2022', xlabel='Date (Yr)', ylabel='Closing Daily Value')
```

```python
# The next step is to create a new column called 'Daily Return'

SP_500['Daily Return'] = (SP_500['Close']/ SP_500['Close'].shift(1)) -1
SP_500.replace(np.NaN, '0')

list(SP_500)
```

```python
SP_500.tail(10)
```

```python
# Plotting S&P500 returns shows us much more than the closing price across the 32 years.

SP_500['Daily Return'].plot(title='S&P 500 Daily Returns', xlabel='Date (Yr)', ylabel='Daily Returns')
```

```python
# However plotting the Daily Return % into a histogram allows us to determine if these daily returns are normally distributed

sb.displot(SP_500['Daily Return'].dropna(), bins=250, color='red', height=4, aspect=23/6)
plt.title('Distribution of S&P500 Daily Returns 1990-2022')
plt.xlim(-0.045, 0.045)
plt.show()
```

```python
# Below is introducing conditional statements to determine the movement of the S&P500 over the period of 1990-2022

SP500_DailyMovements = SP_500['Daily Return']
NegativeDays5pct = len(list(filter(lambda x: (x < -0.05), SP500_DailyMovements)))
PositiveDays5pct = len(list(filter(lambda x: (x >= 0.05), SP500_DailyMovements)))

if True: 
    if NegativeDays5pct == PositiveDays5pct:
        print("The S&P has closed up and down 5% or more the same number of times over 32 years")
    elif PositiveDays5pct > NegativeDays5pct:
        print("There are more positive days of 5% or more than down days of 5%")
    elif NegativeDays5pct <= PositiveDays5pct: 
        print("The negative days of 5% or less or equal to the positive days on the S&P")
    else:
        print("The S&P500 has closed down 5% more frequently between 1990-2002 than the index closed up 5%")
```

```python
# Here are the results of the conditional statements and yes we can see S&P500 has dropped 5% or more 21 times since 1990.

SP500_DailyMovements = SP_500['Daily Return']

Negative_5pct = len(list(filter(lambda x: (x < -0.05), SP500_DailyMovements)))
Positive_5pct = len(list(filter(lambda x: (x >= 0.05), SP500_DailyMovements)))

print("The number of days we saw a growth of 5% or more in the S&P500 =", Positive_5pct)
print("The number of days we saw a decline of 5% or more in the S&P500 =", Negative_5pct)
```

```python
# However, the growth of the S&P500 drastically outweighs the days where the index dropped when we reduce the variable to 0.

SP500_DailyMovements = SP_500['Daily Return']

NegativeDays = len(list(filter(lambda x: (x < 0), SP500_DailyMovements)))
PositiveDays = len(list(filter(lambda x: (x >= 0), SP500_DailyMovements)))

print("Positive closing days in the S&P500 between Jan, 1990 & May, 2022 =", PositiveDays)
print("Negative closing days in the S&P500 between Jan, 1990 & May, 2022 =", NegativeDays)
```

```python
# Finally lets add one more to show daily returns against the annualised returns

SP_500['Daily Return'] = SP_500['Adj Close'].pct_change()
SP_500.dropna(inplace=True)

def annualize(r, n, p, t=1):
  return round(p * (1 + r/n)**(n*t),2)/100

SP_500["Annualized Return"] = SP_500["Daily Return"].apply(annualize, p=100, n=252)

SP_500.head(10)
```
